```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel("This is a x lebel")
plt.ylabel("This is a y lebel")
plt.title("This is a x and y")
```

```python
plt.scatter(x,y ,c='r')
plt.xlabel("This is a x lebel")
plt.ylabel("This is a y lebel")
plt.title("This is a x and y")
```

```python
plt.scatter(x,y ,c='#33F9FF')
plt.xlabel("This is a x lebel")
plt.ylabel("This is a y lebel")
plt.title("This is a x and y")
```

```python
plt.scatter(x,y ,c='#33F9FF',alpha=.8)
plt.xlabel("This is a x lebel")
plt.ylabel("This is a y lebel")
plt.title("This is a x and y")
```

```python
plt.scatter(x,y ,c='#33F9FF',alpha=.8,marker="X")
plt.xlabel("This is a x lebel")
plt.ylabel("This is a y lebel")
plt.title("This is a x and y")
```

```python
data=[1,2,3,4,5,6]
data1=[2,3,4,5,6,7]
```

```python
plt.plot(data,data1)
plt.xlabel("this is a x lebel ",c='g')
plt.ylabel("this is a y lebel ",c='y')
plt.title("this is a title" , c='b')
plt.grid()
plt.figure(figsize=(5,6))
```

```python
plt.scatter(data,data1)
plt.xlabel("this is a x lebel ",c='g')
plt.ylabel("this is a y lebel ",c='y')
plt.title("this is a title" , c='b')
plt.grid()
plt.figure(figsize=(5,6))
```

```python
plt.bar(data,data1)
plt.xlabel("this is a x lebel ",c='g')
plt.ylabel("this is a y lebel ",c='y')
plt.title("this is a title" , c='b')
plt.grid()
plt.figure(figsize=(5,6))
```

```python
data2=[1,2,1,1,2,2,3,4,3,5,6,6,7,5,4]
```

```python
plt.hist(data2)
plt.xlabel("this is a x lebel ",c='b')
plt.ylabel("this is a y lebel ",c='y')
plt.title("this is a title" , c='b')
plt.grid()
plt.figure(figsize=(5,6))
plt.savefig("hist.png")
```

```python
#plot 3d figure
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
ax.scatter(x,y,z)
plt.xlabel("this is a x lebel ",c='b')
plt.ylabel("this is a y lebel ",c='y')
plt.title("this is a title" , c='b')
plt.grid()
plt.figure(figsize=(5,6))
plt.show()
```
