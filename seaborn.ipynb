```python
tips=snb.load_dataset('tips')
```

```python
tips
```

```python
snb.scatterplot(x=tips.total_bill,y=tips.size)
```

```python
snb.scatterplot(x=tips.total_bill,y=tips.smoker)
```

```python
snb.scatterplot(x=iris.sepal_length ,y=iris.sepal_width)
```
